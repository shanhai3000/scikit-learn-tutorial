# 核密度估计分布图(平滑直方图)/Kernel Density Estimate

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris
print("Setup Complete")
```

```python
iris_data = load_iris()
columns = iris_data.feature_names + ['target']
iris_df = pd.concat([pd.DataFrame(iris_data.data), pd.DataFrame(iris_data.target)], axis=1)
iris_df.columns = columns
print(iris_df.head())
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(data=iris_df['petal length (cm)'], shade=True)
```

# 二维核密度分布图
- 必须是相同规模的数据，才能形成二维contour plot

```python
sns.jointplot(x=iris_df['petal length (cm)'], y=iris_df['sepal width (cm)'], kind='kde')
```
