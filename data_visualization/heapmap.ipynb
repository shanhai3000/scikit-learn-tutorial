```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
print(data.target)
data_fm = pd.concat([pd.DataFrame(data.data), pd.DataFrame(data.target)], axis=1)
print(data_fm.shape)
```

```python
_data = data_fm.iloc[:20]
plt.figure(figsize=(20,20))
sns.heatmap(data=_data, annot=True)
```

```python
a = [
    [1,0,2,1]
    ,[1,1,6,2]
    ,[0,2,0,1]
]

b = [
    2,3,4
]

pd2 = pd.concat([pd.DataFrame(a), pd.DataFrame(b)], axis=1)
print(pd2)

sns.heatmap(data=pd2, annot=True)
```
