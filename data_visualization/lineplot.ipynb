# 折线图

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris

iris_data = load_iris()
# data, target, target_name
```

```python
plt.figure(figsize=(12,6))
plt.title("iris flowers")
sns.lineplot(data=iris_data.data)
```

```python
iris_df = pd.concat([pd.DataFrame(iris_data.data), pd.DataFrame(iris_data.target)], axis=1)
print(iris_df)
print(iris_data.target_names)
```

```python
player_time_data = pd.read_csv("../dataset/player_time.csv", index_col='match_id')
```

```python
match_i_data = player_time_data.iloc[player_time_data.index == 0]
index = match_i_data.times
columns = ['gold_t_0', 'lh_t_0', 'xp_t_0', 'gold_t_1', 'lh_t_1', 'xp_t_1',
       'gold_t_2', 'lh_t_2', 'xp_t_2', 'gold_t_3', 'lh_t_3', 'xp_t_3',
       'gold_t_4', 'lh_t_4', 'xp_t_4', 'gold_t_128', 'lh_t_128', 'xp_t_128',
       'gold_t_129', 'lh_t_129', 'xp_t_129', 'gold_t_130', 'lh_t_130',
       'xp_t_130', 'gold_t_131', 'lh_t_131', 'xp_t_131', 'gold_t_132',
       'lh_t_132', 'xp_t_132']
gold_columns = [i for i in columns if i.find('gold') >= 0]
match_i_data.index= index
#You have to provide both hue="B" and style="B", and then provide a dictionary to markers={...} (you actually passed a list containing a dictionary in your code)
sns.lineplot(data=match_i_data[gold_columns],  dashes=False)
print(match_i_data.shape)
```
