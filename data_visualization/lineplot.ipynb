```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris

iris_data = load_iris()
# data, target, target_name
```

```python
plt.figure(figsize=(12,6))
plt.title("鸢尾花")
sns.lineplot(data=iris_data.data)

```

```python
iris_df = pd.concat([pd.DataFrame(iris_data.data), pd.DataFrame(iris_data.target)], axis=1)
print(iris_df)
print(iris_data.target_names)
```
